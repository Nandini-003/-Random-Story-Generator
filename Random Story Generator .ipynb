```python
import random

# Define the story templates with placeholders
story_templates = [
    "Once upon a time, in a [adjective] land, there was a [noun] who loved to [verb].",
    "The [noun] was very [adjective] and spent its days [verb] in the [place].",
    "One day, a [adjective] [noun] decided to [verb] and find a new [noun].",
    "In a [adjective] forest, a [noun] met a [adjective] [noun] who wanted to [verb].",
    "A [adjective] [noun] went to the [place] to [verb] and discovered a [noun]."
]

# Define lists of words for each part of speech
adjectives = ["happy", "mysterious", "ancient", "brave", "furry"]
nouns = ["dragon", "princess", "wizard", "castle", "elf"]
verbs = ["fly", "dance", "sing", "explore", "run"]
places = ["mountain", "forest", "village", "river", "desert"]

def generate_story():
    # Choose a random story template
    template = random.choice(story_templates)

    # Replace placeholders with random words
    story = template.replace("[adjective]", random.choice(adjectives))
    story = story.replace("[noun]", random.choice(nouns))
    story = story.replace("[verb]", random.choice(verbs))
    story = story.replace("[place]", random.choice(places))

    return story

def main():
    print("Welcome to the Random Story Generator!")
    while True:
        print("\nGenerating a random story...")
        print(generate_story())
        cont = input("\nWould you like to generate anoyesther story? (yes/no): ")
        if cont.lower() != 'yes':
            print("Goodbye!")
            break

if __name__ == "__main__":
    main()
```
